```python
import pandas as pd
import re
```

```python
# Load data into a pandas df
df_true = pd.read_csv('../data/raw_data/True.csv')
df_true
```

```python
# remove duplicate items
df_true.drop_duplicates(inplace=True)
df_true
```

```python
# Create new df with text column only
df_true_small = df_true.drop(columns=['title', 'subject', 'date'], axis=0)
df_true_small.head()
```

```python
# check for null values
df_true_small.isnull().sum()
```

```python
#
df_true_small = df_true_small.drop_duplicates()
df_true_small.duplicated().sum()
```

```python
df_true_small['is_fake'] = 0
```

```python
df_true_small.reset_index(drop=True, inplace=True)
```

```python
def remove_city_name(text):
    pattern = r'(?:[A-Z]+(?:[ \/][A-Z]+)* )?\(Reuters\) - '
    return re.sub(pattern, '', text)
```

```python
df_true_small['text'] = df_true_small['text'].apply(remove_city_name)
df_true_small.head()
```

```python
df_true_clean = df_true_small.copy()
df_true_clean
```

```python
df_true_clean.to_csv('../data/cleaned_data/true_cleaned.csv', index=False)
```
